```python
dataset['Gender'].replace('Female',0, inplace=True)
dataset['Gender'].replace('Male',1, inplace=True)
```

```python
dataset['family_history_with_overweight'].replace('no',0, inplace=True)
dataset['family_history_with_overweight'].replace('yes',1, inplace=True)
```

```python
dataset['SMOKE'].replace('no',0, inplace=True)
dataset['SMOKE'].replace('yes',1, inplace=True)
```

```python
dataset['NObeyesdad'].replace('Insufficient_Weight',0, inplace=True)
dataset['NObeyesdad'].replace('Normal_Weight',1, inplace=True)
dataset['NObeyesdad'].replace('Overweight_Level_I',2, inplace=True)
dataset['NObeyesdad'].replace('Overweight_Level_II',2, inplace=True)
dataset['NObeyesdad'].replace('Obesity_Type_I',3, inplace=True)
dataset['NObeyesdad'].replace('Obesity_Type_II',3, inplace=True)
dataset['NObeyesdad'].replace('Obesity_Type_III',3, inplace=True)
```

```python
dataset = dataset.drop(["MTRANS","FAVC","FCVC","NCP","CAEC","CH2O","SCC","FAF","TUE","CALC"], axis=1)
dataset
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 6].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=40)
xtrain.shape, xtest.shape
```

```python
norm = StandardScaler()
xtrain = norm.fit_transform(xtrain)
xtest = norm.transform(xtest)
```

```python
from sklearn.svm import SVC
from sklearn import metrics

svc=SVC()
svc.fit(xtrain,ytrain)
y_pred=svc.predict(xtest)

print('Accuracy Score:')
print(metrics.accuracy_score(ytest,y_pred))
```

```python
my_pred = svc.predict([[0,22,1.54,89,0,1]]) #Gender(0-M,1-F),age,height,weight,family history(),smoking habit
if my_pred==1:
  print("You are Normal Weight")
elif my_pred==0:
  print("You are Insufficient in Weight")
elif my_pred==2:
  print("You are Over Weight")
else:
  print("You have Obesity")
```
